Scipy nedir? bilimsel ve teknik hesaplamalar için kullanılan açık kaynaklı bir Python kütüphanesidir. Temel veri yapılarını oluşturduğu NumPy kütüphanesi üzerine kurulmuştur. Optimizasyon, integral, diferansiyel denklem çözümleri ve istatistik gibi gelişmiş matematiksel işlemleri gerçekleştirmek için kullanılır

```python
from scipy import optimize
from scipy import stats
from scipy.optimize import fsolve

def denklem(x):
    return x**2 - 4*x + 3

baslangic_tahmini = 0
cozum = fsolve(denklem, baslangic_tahmini)

print("Denklemin kökü:", cozum[0])
```
